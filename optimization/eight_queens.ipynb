```python
import eight_queens as eq
import matplotlib.pyplot as plt
```

```python
for z in range(0,5):
    run = eq.run_ga_with_metrics(40, 40, 3, 0.7, 2)
    best = run[0]
    metrics = run[1]
    # print(best)
    print(eq.evaluate(best))
    # matrix = list()
    # for i in range(0,8):
    #     row = [0]*8
    #     row[best[i]-1] = 1
    #     matrix.append(row)
    # for i in range(0,8):
    #     for j in range(0,8):
    #         print(str(matrix[j][7-i])+" ", end='')
    #     print()

    x = list()
    genAvg = list()
    genMax = list()
    genMin = list()

    count = 1
    for gen in metrics:
        # print(gen)
        sum = 0
        for ev in gen:
            sum += ev
        avg = sum/len(gen)
        # print(gen[0], avg, gen[-1])
        genAvg.append(avg)
        genMax.append(gen[-1])
        genMin.append(gen[0])
        x.append(count)
        count += 1
    plt.plot(x, genAvg, label = "average")
    plt.plot(x, genMax, label = "max")
    plt.plot(x, genMin, label = "min")
    plt.legend()
    plt.show()
```
