```python
import eight_queens_extra as eq
import matplotlib.pyplot as plt
runs = 1
runsSum = 0
for z in range(0,runs):
    run = eq.run_ga_with_metrics(80, 400, 3, 0.7, 30)
    best = run[0]
    metrics = run[1]
    # print(best)
    bestValue = eq.evaluate(best)
    runsSum += bestValue
    print(bestValue)

    x = list()
    genAvg = list()
    genMax = list()
    genMin = list()

    count = 1
    for gen in metrics:
        # print(gen)
        sum = 0
        for ev in gen:
            sum += ev
        avg = sum/len(gen)
        # print(gen[0], avg, gen[-1])
        genAvg.append(avg)
        genMax.append(gen[-1])
        genMin.append(gen[0])
        x.append(count)
        count += 1
    plt.plot(x, genAvg, label = "average")
    plt.plot(x, genMax, label = "max")
    plt.plot(x, genMin, label = "min")
    plt.xlabel('Generation')
    plt.ylabel('Conflicts')
    plt.legend()
    plt.show()

print("Average:")
print(runsSum/runs)
```
